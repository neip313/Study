- 쉘 스크립트 상단에 반드시 기입해야 할 것
```
#!/bin/bash
``` 
- 예제 코드
```
test="hello world"
```
- echo $test (=붙여쓰기)
 
- 저장하고 종료 등 먼저 ESC 누르고  

```
:w 저장 :wq 저장후 종료
:q 나가기
:q! 무시하고 나가기
```

```
./temp.sh
```
- 코드 쳤을 때 hello world가 나오면 성공
 
---
