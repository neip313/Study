# 구글 빅쿼리

## 1. 데이터 처리 아키텍처 
- 구글 빅쿼리는 쿼리 엔진이 내장된 서버리스 서비스로 확장설이 높은 데이터 웨어하우스
- 빅쿼리의 가장 큰 특징은 규모와 속도
- 서버리스에 내재한 단순함과 에드혹(ad hoc)쿼리 덕분에 새로운 방식을 작업할 수 있어서 인프라를 운영할 필요가 없다
- 빅쿼리는 민첩하며 개방적인 문화로의 전환을 가능케 하는 기술적 수단을 제공해 혁신 속도를 높이는 데 큰 역할

### 1) 관계형 데이터베이스 관리 시스템
- OLTP(Online Transaction Processing)데이터베이스의 주요 장점 중 하나는 SQL(Structured Query Language)을 지원한다는 점

**OLTP 데이터베이스의 장점과 단점**  
1. SQL은 단순히 데이터베이스 컬럼에서 원시 데이터를 가져오는 것뿐만 아니라 그 이상의 작업도 가능 
2. SQL을 사용해 얻을 수 있는 가장 주된 장점은 원하는 데이터만 명시하면 데이터베이스 소프트웨어가 쿼리를 실행하는 최적의 방법을 스스로 찾아낸다는 점
3. OLTP 데이터베이스는 데이터 기록과 동시에 읽을 수 있도록 데이터 일관성을 확보하는 데 최적화
4. OLTP 데이터베이스는 이처럼 전체 데이터셋을 훑어야 하는 에드혹 쿼리 실행에는 부적합
    - 에트훅 쿼리는 필요에 따라 (성능이 조금 떨어져도)즉석에서 작성해 실행하는 쿼리  

### 2) 맵리듀스 프레임워크
- 키/값 쌍을 처리해 중간 키/값 쌍을 생성하는 map함수, 동일한 중간키와 연관된 모든 중간 값을 병합하는 reduce 함수, 이를 맵리듀스(mapreduce)라 한다.  

### 3) 빅쿼리 : 서비리스, 분산 SQL 엔진 
- 빅쿼리는 서버리스 서비스이므로 인프라를 관리할 필요 없이 쿼리를 실행
- 전체 데이터셋에 대한 집계를 수 초에서 수 분 내에 처리하는 분석을 수행

## 2. 빅쿼리로 작업하기 
- 빅쿼리는 중앙 집중화가 되어 있고 보편성을 가지는 데이터 웨어하우스
- 빅쿼리의 장점은 다른 소스의 데이터셋을 조인하거나 빅쿼리 외부에 저장된 데이터에 쿼리할 때 빛을 발한다. 
- 빅쿼리가 서버리스라는 점은 기술적으로 부서 간에 필요한 데이터를 서로 나누어 유연하게 협업할 수 있음을 의미

### 1) ETL, EL, ELT
- 빅쿼리는 ETL을 매력적인 방법으로 만드는 매우 효율적인 컬럼 스토리지 포맷
- 빅쿼리로 ETL 파이프라인 구축 가능
- 데이터를 데이터 웨어하우스로 로드하는 이유는 네이티브 스토리지에 저장할 때 쿼리 성능이 가장 효율적이기 때문

### 2) 강력한 분석 
- 데이터 웨어하우스는 저장된 데이터로 작업할 수 있는 분석의 종류가 많다는 장점
- 다양한 비즈니스 인텔리전스(BI) 도구로 빅쿼리에 저장된 데이터를 영향력 있는 분석, 시각화, 보고서로 만들 수 있다. 
- 빅쿼리는 두 테이블을 거리나 겹침 조건(overlap criteria)을 기준으로 조인하는 기능을 포함해서 위치 인식(location-aware)쿼리를 지원하는 수많은 공간 함수를 제공하는 확장 기능
- 빅쿼리는 서술적 분석을 할 수 있는 강력한 엔진
- 표준 SQL에 대한 또 다른 빅쿼리 확장 기능은 머신러닝 모델 생성과 배치 예측 작업을 지원
- 빅쿼리에서 데이터를 내보내지 않고도 모델을 학습하고 예측할 수 있다.
- 빅쿼리는 현황 분석뿐만 아니라 예측 분석도 지원하는 데이터 웨어하우스
- 데이터 웨어하우스는 다양한 유형의 데이터를 저장
- 빅쿼리는 숫자 및 텍스트 컬럼뿐만 아니라 지리 공간 데이터와 계층 데이터 등 다양한 유형의 데이터를 저장
- 빅쿼리는 배치 데이터와 스트리밍 데이터 수집 모두 지원 

### 3) 관리의 단순함 
- 빅쿼리의 스토리지는 완전한 관리형이므로 데이터를 적재할 때 스토리지의 종류나 속도 및 비용의 트레이드오프를 생각할 필요가 없다. 
- 빅쿼리는 인텍스가 필요하지 않다. 
- SQL 쿼리는 데이터셋의 모든 컬럼을 필터링하며, 빅쿼리는 필요한 쿼리 계획 및 최적화를 처리
- 쿼리는 자동으로 수천 대의 머신으로 확장되고 병렬로 실행
- 머신 자체가 작업의 여러 단계를 처리하도록 투명하게 프로비저닝
- 인프라스트럭처를 직접 구축하지 않아도 되므로 보안을 신경 써야 하는 번거로움도 줄어듬
- 빅쿼리의 데이터는 저장 및 전송 시 자동으로 암호화

## 3. 빅쿼리는 어떻게 만들어졌는가
- 클라우드에 저장된 데이터로 연산을 수행할 때 어떤 언어를 사용하는 것이 가장 좋을까? 
    1. SQL은 개발자뿐만 아니라 여러 사람이 데이터로 질문하고 문제를 해결할 수 있게 해주는 다용도 언어
    2. SQL은 '관계상 완전'하므로 데이터에 대한 모든 연산을 SQL을 사용해 수행 
    3. 항상 SQL은 종료(terminate)되므로 '튜링 완전'하지 않은 언어

## 4. 빅쿼리는 어떻게 구현할 수 있었을까
빅쿼리를 성공적이고 특별하게 만든 일부 구성 요소를 알아보자 

### 1) 컴퓨팅 및 스토리지 분리 
- 빅쿼리처럼 스토리지와 컴퓨팅을 분리하면 더 이상 데이터가 필요하지 않는 경우가 아니라면 데이터를 파기(삭제)하지 않아도 된다. 
- 데이터를 영구적으로 보관하고 언제든지 필요할 때 접근할 수 있다는 것은 매우 강력한 장점
- 빅쿼리는 대량의 하드웨어 리소스 풀을 관리하는 멀티테넌트 시스템이므로 사용자별 또는 쿼리별로 슬롯을 분리할 수 있다. 
- 프로젝트 또는 조직을 위해 하드웨어를 예약하거나 공유 주문형풀에서 쿼리를 실행
- 빅쿼리는 컴퓨팅과 스토리지가 분리되어 있어 광범위한 고객 요구를 수용

### 2) 스토리지 및 네트워킹 인프라 
- 빅쿼리의 쿼리는 분산 파일 시스템의 회전식 디스크에서 처리
- 빅쿼리는 구글의 고유 시스템인 콜로서스 파일 시스템과 주피터 네트워킹에 의존해서 컴퓨팅 클러스터의 실제 위치에 관계없이 데이터를 신속하게 쿼리 
- 컴퓨팅 및 스토리지를 동일 위치에 배치하지 않아도 빅쿼리가 작동
- 고속 네트워킹 패브릭은 디스크에서 데이터 읽기와 쿼리 단계의 셔플에 특히 유용
- 네트워킹 인프라는 속도뿐만 아니라, 대역폭도 동적으로 프로비저닝 
- 빅쿼리는 효과적인 처리량과 효율적인 데이터 형식 및 스토리지를 결합해 페타바이드 규모의 테이블을 몇분 만에 쿼리할 수 있는 기능을 제공 

### 3) 관리형 스토리지
- 빅쿼리는 기반 데이터 형식과 파일 크기를 추상화함으로써 사용자에게 노출하지 않아 원활한 환경을 제공하며 쿼리도 항상 빠르게 처리
- 빅쿼리가 스토리지를 자체적으로 관리해서 생기는 장점
    - 최종 사용자에게 투명한 방식으로 빅쿼리가 계속해서 빨라질 수 있다는 점 
- 빅쿼리가 모든 스토리지를 관리하므로 사용자는 백업 또는 복제를 걱정할 필요가 없다. 
- 빅쿼리의 스토리지 관리 시스템은 업그레이드, 복제, 백업 및 복원에 이르는 모든 작업을 투명하게 자동으로 처리
- 빅쿼리 작업은 ACID다. 즉 모든 쿼리가 완전히 커밋되거나 전혀 커밋되지 않는다. 
- 관리형 스토리지에 있는 데이터를 읽을 때 빅쿼리는 유효성 검사 오류가 발생하지 않도록 보장
- 관리형 스토리지의 한 가지 단점은 다른 프레임 워크를 사용해 데이터에 직접 엑세스하고 처리하기가 더 어렵다는 점
- 빅쿼리는 데이터를 읽을 수 있는 구조화된 병렬 API를 제공해 이 문제를 해결 

### 4) 구글 클라우드 플랫폼과 통합 
- 구글 클라우드는 소수의 우수한 제품이 서로 긴밀하게 통합되는 '책임 분리'설계 원칙을 따른다. 

### 5) 보안 및 규정 준수 
- 빅쿼리와 GCP와 통합은 다른 제품과의 상호 운용성을 넘어서고 있다. 플랫폼에서 제공하는 크로스 커팅 기능은 일관된 보안 및 규정 준수를 제공
- 빅쿼리에 데이터를 수집하려면 EL 파이프라인(일반적으로 로그 파일의 주기적 로드에 사용됨), ETL 파이프라인(일반적으로 데이터가 풍부하거나 품질 관리가 필요한 경우에 사용됨)또는 ELT 파이프라인(일반적으로 탐색 작업에 사용됨)을 사용할 수 있다. 
- 빅쿼리는 OLAP(데이터 분석)를 위해 설계되었으며 SQL:2011의 모든 기능을 지원 
- 빅쿼리는 컬럼 기반 스토리지 사용, 중첩 및 반복 필드 지원, 컴퓨팅과 스토리지의 분리 등 구글이 발표한 논문에서 소개한 혁신적인 엔지니어링 아이디어를 기반으로 확장성과 속도를 달성
- 빅쿼리는 빅데이터 분석 도구의 GCP 생태계의 일부인 인프라스트럭처 요소(보안, 모니터링 및 로깅 등) 데이터 처리 및 머신러닝(스트리밍, 클라우드 DLP 및 AutoML 등)과 긴밀하게 통합
