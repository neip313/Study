## 2. 생활 기록부 성적 평균 구하기 
- 계산과정
1. 성적을 모두 더한다. 
2. 학생 수를 구한다. 
3. 성적 총합을 학생 수로 나눈다. 

- 필요한 가정 
    - 성적을 모두 더할 수 있다. 
    - 학생 수를 구할 수 있다. 
    
- **"계산 과정 설명에 사용해도 되는 단어와 표현이 무엇인가?"**    

### 1) "성적을 모두 더한다"와 "학생 수를 구한다"를 이해 못할 때
- 계산과정 
    - 1. 성적을 모두 더한다:
        - 1-1. 성적 총합으로 종이에 0을 쓴다. 
        - 1-2. 생활기록부를 순서대로 헤아리며 개별 성적을 총합에 더해서 덮어쓴다. 
        - 1-3. 종이에 적힌 최종 값을 읽는다. 
    - 2. 학생 수를 구한다: 
        - 2-1. 학생 수로 종이에 0을 쓴다. 
        - 2-2. 다시 생활기록부를 순서대로 헤아리며 학생 수를 1 더해서 덮어쓴다. 
        - 2-3. 종이에 적힌 최종 값을 읽는다. 
    - 3. 성적 총합을 학생 수로 나눈다.     
- 필요한 가정     
    - 성적을 모두 더할 수 있다.
    - 학생 수를 구할 수 있다. 
        - 종이에 숫자를 새로 적을 수 있다. 
        - 종이에 적힌 숫자를 읽을 수 있다. 
        - 생활기록부를 순서대로 헤아릴 수 있다. 

### 2) "생활기록부를 순서대로 헤아린다"를 이해 못할 때
- 계산과정 
    - 1. 성적을 모두 더한다:
        - 1-1. 성적 총합으로 종이에 0을 쓴다. 
        - 1-2. 생활기록부를 순서대로 헤아리며 개별 성적을 총합에 더해서 덮어쓴다. 
            - 1-2-1. 더미 맨 위에 놓은 생활기록부를 가져온다. 
            - 1-2-2. 생활기록부에 적힌 성적을 현재 총합에 더해서 기억한다. 
            - 1-2-3. 종이에 새로운 성적 총합을 덮어 쓴다. 
            - 1-2-4. 생활기록부가 남아 있으면 1-2-1로 돌아간다.
        - 1-3. 종이에 적힌 최종 값을 읽는다. 
    - 2. 학생 수를 구한다: 
        - 2-1. 학생 수로 종이에 0을 쓴다. 
        - 2-2. 다시 생활기록부를 순서대로 헤아리며 학생 수를 1 더해서 덮어쓴다. 
            - 2-2-1. 더미 맨 위에 놓은 생활기록부를 가져온다. 
            - 2-2-2. 현재 학생 수에 1을 더해서 기억한다. 
            - 2-2-3. 종이에 새로운 학생 수를 덮어 쓴다. 
            - 2-2-4. 생활기록부가 남아있으면 2-2-1로 돌아간다. 
        - 2-3. 종이에 적힌 최종 값을 읽는다. 
    - 3. 성적 총합을 학생 수로 나눈다.     
- 필요한 가정     
    - 성적을 모두 더할 수 있다.
    - 학생 수를 구할 수 있다. 
        - 종이에 숫자를 새로 적을 수 있다. 
        - 종이에 적힌 숫자를 읽을 수 있다. 
        - 생활기록부를 순서대로 헤아릴 수 있다. 
            - 더미 맨 위에 놓은 생활기록부를 가져올 수 있다. 
            - 생활기록부에 적힌 성적을 읽을 수 있다.

## 4. 계산 설계하기

### 1) 계산 = 무엇 + 어떻게 
- '무엇'과 '어떻게'는 똑같이 중요함

### 2) 생활기록부 성적 평균 구하기

### 3) 연산을 명령으로부터 구분하는 이유 
- 연산 : 다시 수행해도 결과가 똑같음 
    - 전체 설계가 덜 복잡 
- 명령 : 수행할 때마다 결과가 달라짐 
    - 전체 설계가 복잡해짐
