# 1. 계산 자료 


## 1) 기본 자료 

### 1-1. 더 이상 쪼갤 수 없는 계산 자료 
- "성적 총합은 1729이다"
- "생활기록부가 아직 남아 있다"
- "생활기록부가 이제 남아 있지 않다"
- "가장 가까운 도시는 부산이다"
- "시험이 끝나는 날은 수요일이다"

### 1-2. Python 언어의 기본자료 
- 정수 (1728, 0, -183)
- 논리값 (True, False)
- 문자열 ('안녕하세요', "안녕하세요" 'hello world', '', ' ', '""', "''")

### 1-3 문자열 해석하기 
- 문자열이 표현하는 단어와 문장을 프로그램으로 해석하지 않음 

### 1-4 기본자료 = 프로그램 = 계산 결과 
- 기본 자료 실행 
    - "정수 1729를 계산한다"
    
- 기본 자료 계산 = 기본 자료 기억하기     

## 2) 기본 자료로 표현하기 

### 2-1 성적
- 정수로 표현 
- 문자열로 표현 

### 2-2 생년월일 
- 여덟 자리 정수로 표현 
    - 1981년 9월 18일 -> 19810918
- 년, 월, 일을 정수로 표현 
    - 1981년 9월 18일 -> 1981, 9, 18
- 문자열로 표현 
    - 1981년 9월 18일 -> '19810918'

### 2-3 성별 
- 문자열로 표현 
    - '남', '여'
- 논리값으로 표현 
    - True or False
- 정수로 표현 
    - 1, 0, -1

### 2-4 거리 
- 단위 정하고 정수로 표현
    - 40, 40000
- 문자열로 표현     
    - '40km'

### 2-5 요일
- 문자열로 표현 
    - '월요일', '화요일'
- 정수로 표현 
    - 월 = 1, 화 = 2 ....  

## 3) 여러곱 - 계산 자료의 '그리고' 

### 3-1. 접속사 '그리고' 
- 생활기록부에서 이름 그리고 성적을 뽑아낸다. 
- 지도에서 출발 도시 그리고 도착 도시를 골라서 거리를 잰다. 

- 첫째 생활기록부에는 이름 홍길동 그리고 성적 85점이 적혀있다.
- 국어 그리고 영어 그리고 수학은 공통 수강생이 있다. 

### 3-2 프로그래밍 언어에서 '그리고'
- 접속사 대신 특유의 문법을 씀 
- c언어(구조체), python(여러곱), Haskell언어(여러곱), Scala언어(여러곱)


### 3-3 Python 언어 여러곱
- 괄호 사이에 계산 자료 나열하고 쉼표로 분리 
- 각 계산 자료 = 여러곱 원소 
    - (계산 자료1, 계산 자료2, ---, 계산 자료n)
- 사람 언어("정수 1729 그리고 단어 Hello그리고 참") -> Python 언어(1729, 'Hello', True)    

### 3-4 여러곱의 특수한 형태 
- 계산 자료 하나로 엮은 여러곱 
    - (계산 자료, )
    
- 원소가 없는 여러곱 
    - ()

### 3-5 여러곱의 특징 
- 여러곱 = 프로그램 
- 더 큰 여러곱을 만들 때 쓸 수 있음 


## 4) 여러곱으로 표현하기 

### 4-1. 주소 
- 행정 구역 단위별로 엮음 
    - 서울시 중구 명동 -> ('서울시', ;중수', '명동')
   

### 4-2. 생년월일 
- 년, 월, 일 표현하는 정수를 엮음 

### 4-3 거리 
- 단위와 거리 값을 따로 엮음
    - 40km -> ('km', 40)

## 5. 여러합 - 계산 자료의 '또는'

### 5-1 접속사 '또는'
- 생활기록부를 찾으려면 이름 **또는** 생년월일이 필요하다. 
- 킬로미터 거리 **또는** 미터 거리가 나온다. 
- 시험은 월요일 **또는** 화요일 **또는** 수요일에 시작한다. 

### 5-2 이름 또는 생년월일 
- 이름 집합과 생년월일 집합의 합집합을 의미 
- 이름 또는 생년월일 집합이 필요하다. = 이름 집합과 생년월일 집합의 합집합의 원소가 필요하다. 

### 5-3 여러합 
- 접속사 '또는'을 써서 정의한 합집합
- 개별 원소는 종류와 내용을 포함 

- 이름 또는 생년월일 
    - 종류 = 이름, 내용 = 홍길동
    - 종류 = 생년월일, 내용 = 1981년 9월 18일
- 킬로미터 거리 또는 미터 거리 
    - 종류 = 킬로미터, 내용 = 40 
    - 종류 = 미터, 내용 = 40000
- 월요일 또는 화요일 또는 수요일 
    - 종류 = (월, 화, 수)요일, 내용은 따로 없음

### 5-4 프로그래밍 언어에서 여러합
- 프로그래밍 언어마다 다름
- Python 언어에는 여러합의 개념이 없음 

### 5-5 Python 언어에서 여러합 흉내내기 
- 종류와 내용 쌍으로 이루어진 여러곱으로 여러합 원소를 표현 
- 내용이 없으면 종류만으로 이루어진 여러곱으로 표현 
    - 여러합 원소 = (종류, 내용) 
    - 여러합 원소 = (종류, )
- 이름 또는 생년월일  

사람언어|Python 언어
:-|:-
종류=이름, 내용 = 홍길동|('이름', '홍길동')
종류 = 생년월일, 내용 = 1981년 9월 18일|('생년월일',(1981, 9, 18))

### 5-6 여러합 원소 = (종류, 내용)
- 킬로미터 거리 또는 미터 거리  

사람언어|Python 언어
:-|:-
종류=킬로미터, 내용 = 40|('km', 40)
종류 = 미터, 내용 = 40000|('m',40000)
- 월요일 또는 화요일 또는 수요일  

사람언어|Python 언어
:-|:-
종류=월요일, 내용은 따로 없음|('월요일',)
종류=화요일, 내용은 따로 없음|('화요일',)
종류=수요일, 내용은 따로 없음|('수요일',)
