# 8장 기본 명령어

## SELECT
- KEYWORD : 테이블 검색 


## 8.2 WHERE 
- SELECT 명령을 이용해 지정한 테이블 에서 조건에 맞는 데이터를 검색하는 데 사용


## 8.3 AND
- 지정한 테이블에서 나열한 조건을 모두 만족하는 데이터를 검색하고, 하나라도 만족하지 않으면 검색하지 않는다. 


## 8.4 OR 
- KEYWORD : 하나라도 만족하면 검색하기


## 8.5 BETWEEN .. AND
- KEYWORD : 범위에 만족하는 조건 검색 


## 8.6 비교 연산자
- KEYWORD : 크다, 작다, 같다, 다르다, 조건 검색


## 8.7 LIKE
- KEYWORD : 포함된 문자로 검색하기 


## 8.8 IN
- KEYWORD : 다중 데이터 조건 검색
- IN 구문은 OR 조건과 같은 결과


## 8.9 OTHER BY 
- KEYWORD : 데이터 정렬 방식 지정 


## 8.10 GROUP BY 
- KEYWORD : 최소, 최대, 합계, 평균값 등의 취득을 위한 그룹화


## 8.11 DISTINCT 
- KEYWORD : 출력 필드의 중복 없애기


## 8.12 JOIN
- KEYWORD : 테이블의 연관 정보 검색하기 


## 8.13 CASE 
- KEYWORD : 조건에 따른 처리와 값 설정하기


## 8.14 ROWNUM
- KEYWORD : 행 번호 구하기 


## 8.15 NULL
- KEYWORD : 값이 설정되어 있지 않음 


## 8.16 DUAL 
- KEYWORD : 연산 값 출력용 1행 테이블 


## 8.17 UPDATE 
- KEYWORD : 테이블 항목 값 변경 


## 8.18 INSERT 
- KEYWORD l 테이블에 데이터 입력


## 8.19 MERGE
- KEYWORD : 테이블의 내용 삽입 or 변경 처리


## 8.20 DELETE
- KEYWORD : 테이블의 데이터 삭제


## 8.21 COMMIT
- KEYWORD : 데이터베이스 변경 내용을 반영


## 8.22 ROLLBACK
- KEYWORD : 데이터베이스 변경 내용을 취소


## 8.23 트렌잭션의 이해
- KEYWORD : 여러 개의 단위 작업 배치 시 처리 방법


## 8.24 서브쿼리(SUB QUERY)의 이해 
- KEYWORD : SELECT 구문 내 SELECT 작업


## 8.25 EXISTS
- KEYWORD : 다중 데이터 조건 검색


## 8.26 검색 조건을 입력해서 SQL을 실행 
- KEYWORD : 변수로 실행


## 8.27 주석 처리 
- KEYWORD : SQL 구문 내 미실행 문장
