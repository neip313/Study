# 1장_데이터베이스의 기본개념

## 1.1 SQL;
- SQL(Structured Query Language)은?
    - 데이터베이스를 관리하고 조작하기 위해 특화된 질의 언어
- SQL을 배우는 목적
    - 유형/무형의 값을 데이터베이스 내 규격화된 장소에 저장
    - 저장한 데이터를 분석하고 가공해서 정보를 생성

## 1.2 데이터(Data)

## 1.3 필드

## 1.5 테이블
- 테이블(Table)은 연관된 레코드의 모음
- 테이블 출력


## 1.6 데이터베이스와 데이터베이스 관리 시스템 
- 데이터베이스 : 업무단위로 구분된 테이블의 그룹
- 데이터베이스 구성 

- 데이터베이스 관리 시스템(Database Management System : DBMS) : 데이터베이스의 관리 및 조작을 지원하는 프로그램

- DBMS를 사용한 데이터 베이스 제어 방법


## 1.7 SQL 명령의 종류 
- DDL (Data Definition Language)
  - 데이터 정의 언어
  - 테이블 구조를 설정
  - CREATE, ALTER, DROP 명령 등

- DML (Data Manipulation Language)
  - 데이터 조작 언어
  - 테이블 내용을 검색, 입력, 수정, 삭제
  - SELECT, UPDATE, INSERT, DELETE 명령 등

- DCL (Data Control Language)
  - 데이터 제어 언어
  - 테이터베이스를 관리
  - GRANT, REVOKE 명령 등

- TCL (Transaction Control Language)
  - 트랜잭션 제어 언어
  - COMMIT, ROLLBACK 명령
  
### 기본적인 데이터 조작 명령어 
- SELECT : 데이터베이스 내 테이블 내용 검색
- INSERT : 데이터베이스 내 테이블에 내용 입력
- UPDATE : 데이터베이스 내 테이블 내용 수정
- DELETE : 데이터베이스 내 테이블 내용 삭제
### [표 1-5] SELECT 구문의 예
- TB_CUSTOMER 테이블에서 CUTOMER_NM이 ‘김한길’인 데이터를 검색
```
SELECT *
FROM   TB_CUSTOMER
WHERE  CUSTOMER_NM = '김한길';
```

## 1.8 PL/SQL
- 확장 SQL로 함수나 프로시저 내에서 SQL을 프로그램 수준으로 구현 가능

### 지원하는 구문 
- 변수설정,  조건문, 배열처리, 커서 정의, 반복(루프), 예외 처리 등

### [표 1-6] PL/SQL의 반복문
```
 FOR ROW_INFO IN CUR_PNT_INFO(p_year)
    LOOP
        DBMS_OUTPUT.PUT_LINE(ROW_INFO.YEAR || ',' || 
           ROW_INFO.RANK || ',' || ROW_INFO.CUST_CD || ',' ||
           TO_CHAR(ROW_INFO.POINT,'999,999,9999') || ',' || 
           FN_DT_CNVDTFMT(ROW_INFO.LAST_PNT_DT));
        n_tot_cnt := n_tot_cnt + 1;
    END LOOP;
```
- 취득한 값을 반복 처리하여 화면에 출력
