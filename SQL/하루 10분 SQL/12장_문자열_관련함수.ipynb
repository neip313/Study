# 12장 문자열 관련함수 
- 12.1. UPPER( ) : 영문을 대문자로 변환
- 12.2. LOWER( ) : 영문을 소문자로 변환 
- 12.3. SUBSTR( ), SUBSTRB( ) : 문자열 추출 
- 12.4. LENGTH( ), LENGTHB( ) : 문자열 길이 취득 
- 12.5. RPAD( ) : 오른쪽 문자열 채우기 
- 12.6. LPAD( ) : 왼쪽 문자열 채우기 
- 12.7. RTRIM( ) : 오른쪽 공백 제거 
- 12.8. LTRIM( ) : 왼쪽 공백 제거 
- 12.9. TRIM( ) : 양쪽 공백 제거 
- 12.10. INITCAP( ) : 영문 첫 문자 대문자로 변환
