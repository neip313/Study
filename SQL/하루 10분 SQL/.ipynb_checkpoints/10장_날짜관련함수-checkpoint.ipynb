# 10장 날짜관련함수 
- 10.1. SYSDATE : 현재 일시 구하기 
- 10.2. SYSDATE - TO_CHAR( ) : 날짜를 문자로 바꾸기 
- 10.3. SYSDATE - TO_DATE( ) : 날짜형식 문자열을 날짜로 바꾸기 
- 10.4. NEXT_DAY( ) : 지정일 이후의 날짜 구하기 
- 10.5. LAST_DAY( ) : 지정한 특정 월의 마지막 날짜 구하기 
- 10.6. ADD_MONTHS( ) : 지정일의 개월을 더하기 
- 10.7. MONTHS_BETWEEN( ) : 날짜 차이를 개월로 구하기 

## 10.1 SYSDATE 
- KEYWORD : 현재일시


## 10.2 SYSDATE – TO_CHAR( )
- KEYWORD : 날짜형을 문자형으로 변환


## 10.3 SYSDATE – TO_DATE( )
- KEYWORD : 날짜형식을 가진 문자열을 날짜형으로 변환


## NEXT_DAY( )
- KEYWORD : 지정일 이후에 날짜 취득


## 10.5 LAST_DAY( )
- KEYWORD : 지정일 월의 마지막 날짜


## 10.6 ADD_MONTHS( )
- KEYWORD : 지정일에 개월을 더함


## 10.7 MONTHS_BETWEEN( )
- KEYWORD : 두 날짜 사이의 차이를 개월로 취득
